# Analise de Compra

```python
import pandas as pd
import matplotlib.pyplot as plt

df_vendas1 = pd.read_csv('./dataset_clientes_completo.csv')
df_vendas2 = pd.read_csv('./dataset_clientes_completo_2.csv')

print(df_vendas1)
print(df_vendas2)
```

```python
# Mostra o inicio da tabela - Base de Dados
df_vendas1.head(10)
```

```python
# Mostra o final da base de dados - Tabela
df_vendas1.tail(10)
```

```python
# Mostra informacoes da tabela, das series, quantas linhas tem, colunas e valores nulos em cada coluna/serie
df_vendas1.info()
```

```python
df_vendas2.head(10)
```

```python
# Concat junta os dois Data Frames, o ignore_index = True => faz com que o index da junção das duas bases se juntem e contem da maneira correta.
df_completo = pd.concat([df_vendas1,df_vendas2], ignore_index = True)
# um unico dataframe em um so
df_completo
```

```python
df_completo.info()
```

```python
# Filtrar Valores dentro do DataFrame
## filtrando pela idade
df_completo['idade']>18
```

```python
# Filtrando valores 
## filtrando os valores das cidade com a idade mostrando somente a cidade MG e as Pessoas de idade iguais == a 30 anos
df_completo[(df_completo['cidade'] == 'MG') & (df_completo['idade'] == 30)]
```

# LOC and ILOC
## - Tanto o iloc quanto o loc primeiro é linha e depois coluna no [123:123, 1:6]
### loc() 
- Localiza pelo nome da coluna

### iloc() 
- Localiza pelo indice da coluna.

```python
# busca feita pela linha 1997, primeiro linha depois coluna. 
df_completo3.loc[1997]
```

```python
# 
df_completo3.loc[1997, 'renda_mensal']
```

```python
# ILOC
df_completo3.iloc[100:200,1:4]
```
