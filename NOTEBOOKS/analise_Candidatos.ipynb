```python
# Não sei o motivo do erro... Tive que importar novamente a biblioteca pandas e seaborn para continuar a analise!
import matplotlib.pyplot as plt
import seaborn as sns

# Converter colunas relevantes
numericas = ["IDADE", "EXPERIÊNCIA", "PRETENÇÃO SALARIAL", "SALÁRIO DA VAGA"]
df_corr = candidatos[numericas].apply(pd.to_numeric, errors='coerce')

# Mapa de calor da correlação
plt.figure(figsize=(8, 6))
sns.heatmap(df_corr.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlação entre Idade, Experiência, Pretensão e Salário da Vaga")
plt.show()
```

```python
#Experiência × Chance de Contratação
# Ajustar e agrupar por experiência
candidatos["EXPERIÊNCIA"] = pd.to_numeric(candidatos["EXPERIÊNCIA"], errors="coerce")
candidatos["CONTRATADO"] = candidatos["CONTRATADO"].str.lower().str.strip()

# Taxa de contratação por experiência (agrupado em faixas)
bins = [0, 2, 5, 10, 20, 40]
labels = ["0-2", "3-5", "6-10", "11-20", "21+"]

candidatos["FAIXA_EXPERIENCIA"] = pd.cut(candidatos["EXPERIÊNCIA"], bins=bins, labels=labels)

taxa = candidatos.groupby("FAIXA_EXPERIENCIA")["CONTRATADO"].value_counts(normalize=True).unstack().fillna(0)
taxa.plot(kind="bar", stacked=True, colormap="Set2")
plt.title("Taxa de Contratação por Faixa de Experiência")
plt.ylabel("Proporção")
plt.xlabel("Experiência (anos)")
plt.legend(title="Contratado")
plt.show()
```

```python
# Disponibilidade (viagem/mudança) × Contratação
viajem = pd.crosstab(candidatos["DISPONIVEL PARA VIAGEM"], candidatos["CONTRATADO"], normalize="index")
mudanca = pd.crosstab(candidatos["DISPONIBILIDADE DE MUDANÇA"], candidatos["CONTRATADO"], normalize="index")

fig, ax = plt.subplots(1, 2, figsize=(14, 5))
viajem.plot(kind="bar", stacked=True, ax=ax[0], title="Viagem x Contratado", colormap="viridis")
mudanca.plot(kind="bar", stacked=True, ax=ax[1], title="Mudança x Contratado", colormap="plasma")

for a in ax:
    a.set_ylabel("Proporção")
    a.set_xlabel("Disponibilidade")

plt.tight_layout()
plt.show()
```

```python
# Pretensao salarial por estado
candidatos["PRETENÇÃO SALARIAL"] = pd.to_numeric(candidatos["PRETENÇÃO SALARIAL"], errors='coerce')
plt.figure(figsize=(14,6))
sns.boxplot(data=candidatos, x="UF", y="PRETENÇÃO SALARIAL")
plt.title("Pretensão Salarial por Estado (UF)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Taxa de contratação por formação e por área de atuação

# Por formação
form_contratado = candidatos.groupby("FORMAÇÃO")["CONTRATADO"].value_counts(normalize=True).unstack().fillna(0)
form_contratado.plot(kind="bar", stacked=True, figsize=(12,6), title="Taxa de Contratação por Formação", colormap="Accent")

# Por área de atuação
area_contratado = candidatos.groupby("ÀREA DE ATUAÇÃO")["CONTRATADO"].value_counts(normalize=True).unstack().fillna(0)
area_contratado.plot(kind="bar", stacked=True, figsize=(12,6), title="Taxa de Contratação por Área", colormap="Dark2")

plt.tight_layout()
plt.show()
```
