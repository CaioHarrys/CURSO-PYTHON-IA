## Algoritmo Árvore de Decisão

O algoritmo de árvore de decisão é um algoritmo de aprendizagem supervisionada usado para classificação e regressão. É simples e de fácil interpretação, representa decisõe como um diagrama semelhante a uma árvore.

Exemplo:

Vamos prever se alguém vai comprar um produto com base em:
- idade
- renda
  
Suponha os dados:
|Idade|Renda|Comprou?|
|-|-|-|
|25|Alta|Não|
|45|Alta|Sim|
|35|Média|Sim|
|22|Baixa|Não|

A árvore de decisão pode criar divisões assim:

             [Idade > 30?]
              /         \
           Sim           Não
         [Renda?]        Não
       /       \
    Alta       Média/Baixa
    Sim         Sim

Como funciona?

Escolhe a melhor pergunta para separar os dados (com menor "impureza").

Usa critérios como Gini, Entropia ou Erro de classificação.

Repete o processo nos ramos até uma condição final (folha).


Vamos fazer um exemplo para mostrar esse algoritmo na prática



```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Carregar dados
data = load_iris()
X = data.data
y = data.target

# Dividir dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinar modelo
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Avaliar
y_pred = model.predict(X_test)
print("Acurácia:", accuracy_score(y_test, y_pred))

# Visualizar árvore
plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=data.feature_names, class_names=data.target_names, filled=True)
plt.show()
```
