    Tratamento: Preencher ou deletar valores nulos 

```python
df.info()
```

```python
# Identificado os valores nulos na serie uf
df[df['uf'].isna()]
```

```python
# Preenchendo as series com a sua moda

moda = df['uf'].mode()
df['uf'].fillna(moda[0], inplace=True)

modai = df['idioma'].mode()
df['idioma'].fillna(modai[0], inplace=True)

moda2 = df['pcd'].mode()
df['pcd'].fillna(moda2[0], inplace=True)

moda3 = df['status_contratação'].mode()
df['status_contratação'].fillna(moda3[0], inplace=True)

moda4 = df['genero'].mode()
df['genero'].fillna(moda4[0], inplace=True)

moda5 = df['disponível_viagens'].mode()
df['disponível_viagens'].fillna(moda5[0], inplace=True)

moda6 = df['disponível_mudança'].mode()
df['disponível_mudança'].fillna(moda6[0], inplace=True)

df.info()
```

```python
# Mudando o type para datetime

df['data_inscrição'] = pd.to_datetime(df['data_inscricao'], errors='coerce') 
df.drop('data_inscricao',axis=1,inplace=True)
df.info()
```

```python
df.describe()
```

```python
# Preenchendo as colunas com sua media

df['salario_vaga'].fillna(df['salario_vaga'].mean(), inplace=True) 

df['pretenção_salarial'].fillna(df['pretenção_salarial'].mean(), inplace=True)

df['idade'].fillna(df['idade'].median(), inplace=True)

df['anos_experiencia'].fillna(df['anos_experiencia'].mean(), inplace=True)

df['certificacoes'].fillna(df['certificacoes'].mean(), inplace=True)

df.info()
```

```python
# Excluindo a serie nome

df.drop('nome',axis=1,inplace=True)
df.info()
```

```python
# Adicionando outro valor para os nulos nessas series

df['area_atuação'].fillna('Não informado', inplace=True)

df['formação'].fillna('Não informado', inplace=True)
df.info()
```

```python
df['area_atuação'].value_counts()
```

```python
mapeamento_genero = {'Masculino': 0, 'Feminino': 1, 'Outro': 2}
mapeamento_uf = {'BA': 10, 'RS': 11, 'SC': 12, 'DF': 13, 'PE': 14, 'PR': 15, 'CE': 16, 'RJ': 17, 'SP': 18, 'MG': 19}
mapeamento_formação = {'Tecnólogo':20,'Ensino Médio':21,'Graduação':22,'Mestrado':23,'Doutorado':24,'Não informado':25}
mapeamento_area_atuação = {'Tecnologia':30,'Recursos Humanos':31,'Administração':32,'Financeiro':33,'Marketing':34,'Educação':35,'Engenharia':36,'Saúde':37,'Não informado':38}
mapeamento_idioma = {'Nenhum':40,'Alemão':41,'Francês':42,'Inglês':43,'Espanhol':44}

# Converter 'genero'
df['genero_int'] = df['genero'].map(mapeamento_genero)
df['genero_int'] = df['genero_int'].astype('int64')

df['formação_int'] = df['formação'].map(mapeamento_formação)
df['formação_int'] = df['formação_int'].astype('int64')

df['area_atuação_int'] = df['area_atuação'].map(mapeamento_area_atuação)
df['area_atuação_int'] = df['area_atuação_int'].astype('int64')

df['idioma_int'] = df['idioma'].map(mapeamento_idioma)
df['idioma_int'] = df['idioma_int'].astype('int64')

# Converter 'uf' com tratamento para valores não encontrados
def converter_uf(uf):
    return mapeamento_uf.get(uf, -1) # Retorna -1 para UFs não mapeadas

df['uf_int'] = df['uf'].apply(converter_uf).astype('int64')
```

```python
# Excluindo as antigas colunas

df.drop('genero', axis=1, inplace=True)
df.drop('uf', axis=1, inplace=True)
df.drop('formação', axis=1, inplace=True)
df.drop('area_atuação', axis=1, inplace=True)
df.drop('idioma', axis=1, inplace=True)
```

```python
mapeamento_disponível_viagens = {'Sim': 1, 'Não': 0}
mapeamento_pcd = {'Sim': 1, 'Não': 0}

df['disponível_viagens_int'] = df['disponível_viagens'].map(mapeamento_disponível_viagens)
df['disponível_viagens_int'] = df['disponível_viagens_int'].astype('int64')

df['pcd_int'] = df['pcd'].map(mapeamento_pcd)
df['pcd_int'] = df['pcd_int'].astype('int64')
```

```python
df.drop('disponível_mudança_int', axis=1, inplace=True)
df.drop('disponível_viagens', axis=1, inplace=True)
df.drop('pcd', axis=1, inplace=True)
```
