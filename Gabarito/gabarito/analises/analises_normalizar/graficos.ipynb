```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y = [2,3,4,3]
```

# Grafico de Linhas

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, label='Informações')
plt.title('Grafico de linha')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.legend()
plt.show()
```

```python
plt.plot(x,y, label='Informações')
plt.title('Grafico de linha')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.yticks([1,2,3,4])
plt.xticks([1,2,3,4])
plt.legend()
plt.show()
```

```python
plt.plot(x ,y, ls=':', lw='10.0', color='#7356b0')
plt.show()
```

# Grafico de Barras

```python
plt.bar(x, y)
plt.show()
```

```python
plt.bar(x,y, label='Informações', color='green')
plt.title('Grafico de Barras')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y, label='Triangulo', color='red', marker='v')
plt.title('Grafico Scatter')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.legend()
plt.show()
```

```python
figure = plt.figure(figsize=(20,10))

figure.add_subplot(221)  # O numero deve ser de 3 digitos, ele informa a posição do grafico
plt.plot(x,y, label='Informações')
plt.title('Grafico de linha')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.legend()

figure.add_subplot(222) # 1° digito indica a quatidade de linhas, o 2° o n° de colunas e o 3° indica o index onde o grafico vai ficar
plt.bar(x,y, label='Informações', color='green')
plt.title('Grafico de Barras')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.legend()

figure.add_subplot(223)
plt.scatter(x,y, label='Triangulo', color='red', marker='v')
plt.title('Grafico Scatter')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.legend()

plt.show()
```

```python
valore = [3,5,10,3,1,8,7,7,7,10,5,3]
plt.hist(valore, label='Frequencia',color='#3db8b8')
plt.title('Grafico Histograma')
plt.xlabel('eixo x')
plt.legend()
plt.show()
```
